```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_dataset(file_path): 
    return pd.read_csv(file_path) 
df = load_dataset("./data.csv") 
df.head(5)
```

```python
df.tail(5)
```

```python
sns.countplot(data=df, x='Crash Type')
df['Year'].value_counts()
```

```python
df['Crash Type'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
plt.hist(df['Age'])
```

```python
# Scatter plot (Numerical - Numerical) for the relation between two columns
sns.scatterplot(x='Age', y='Year', hue='Crash Type', style='Bus Involvement', size='Month', data=df)
plt.show()
```

```python
# Bar plot (Numerical - Categorical)
sns.barplot(x='Crash Type', y='Age', hue='Bus Involvement', data=df)
plt.show()
```

```python
# Box plot (Numerical - Categorical)
sns.boxplot(x='Crash Type', y='Age', hue='Bus Involvement', data=df)
plt.show()
```

```python
sns.displot(data=df, x='Year', hue='Crash Type', kde=True, multiple="stack")
plt.show()
```

```python
# Pairplot 
sns.pairplot(df, hue='Crash Type')
plt.show()
```
